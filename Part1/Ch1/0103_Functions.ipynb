# 함수(Functions)

$f(x) = x^2$

```python
def my_func(x):
    return x**2

# x_values = []
# y_values = []
# for i in range(-10, 11):
#     x = float(i)
#     x_values.append(x)
#     y_values.append(my_func(x))

x_values = [float(i) for i in range(-10, 11)] # List comprehension
y_values = [my_func(x) for x in x_values]

print(x_values)
print(y_values)
```

```python
import numpy as np

x_values = np.linspace(-10, 10, 21)
y_values = my_func(x_values)
# y_values = x_values ** 2

print(x_values) # x_values.tolist()
print(y_values) # y_values.tolist()
```

```python
import matplotlib.pyplot as plt

plt.plot(x_values, y_values, label="y = x²")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = x²")
plt.legend()

plt.grid(True)
plt.show()
```

### 함수 객체

```python
def func1(x):
    return x ** 2

def func2(x):
    return 10 * x

def func3(x):
    return x + 3

f1 = func1
f2 = func2
f3 = func3

print(f1(3))         # x²
print(f2(f1(3)))     # 10x²
print(f3(f2(f1(3)))) # 10x² + 3
```

```python
x = 3

func_list = [f1, f2, f3]

for f in func_list:
    x = f(x)

print(x)
```
